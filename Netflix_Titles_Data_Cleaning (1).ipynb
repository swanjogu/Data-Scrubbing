## Question 1:

## Give the number of all Movies originating from South Africa

```python
# Filtering  movies from South Africa
south_africa_movies = data[(data['type'] == 'Movie') & (data['country'].str.contains('South Africa', na=False))]
south_africa_movies.shape[0]
```

## The dataset was filtered to include only rows where the type is Movie and the country contains South Africa. The total number of such movies was then counted and it's equals to 51

## Question 3:

## From these South African movies, list only those released in 2021

```python
south_africa_2021 = data[
    (data['type'] == 'Movie') &
    (data['country'].str.contains('South Africa', na=False)) &
    (data['release_year'] == 2021)
]

south_africa_2021[['title','release_year','country']]
```
